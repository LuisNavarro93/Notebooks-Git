## Trajectory profiles

```python
plt.figure(figsize=(16,10))
plt.plot(np.arange(0,max(traj_wellbore["md"])),np.arange(0,max(traj_wellbore["md"])),'--r',alpha=.2,label="Vertical")
plt.plot(traj_wellbore["md"],traj_wellbore["tvd"],label="Actual") ; plt.xlabel("md") ;plt.ylabel("TVD") ; plt.title("MD vs TVD")
plt.legend()
plt.show()
```

As seen from the plot, the wellbore starts deviating before the first 1000mts

```python
sc.plts_traj(traj_wellbore,"incl","tvd","Inclination","invert")
```

The  inclination of a wellbore is the angle between the bit path and the vertical axis; the segments between 500 and 1200, could be explained by a operational deviation (suden increase in angle, pikes), while from 1500 to 2500 a constant angle to in

```python
sc.plts_traj(traj_wellbore,"azi","tvd","Azimuth","invert")
```

The azimuth beeing the angle formed between the magnetic north and the bit trajectory, is a indicator of the Northing-Eastern displacement, later plotted.
